```python
import re

def find_char(x):
    if re.match("N",x) != None:
        a = re.match("N",x).group(0)
        return a
    elif re.match("Y",x) != None:
        y = re.match("Y",x).group(0)
        return y
    else:
        return None
        
find_char(copiadfsharks3["Fatal(Y/N)"][0])
```
