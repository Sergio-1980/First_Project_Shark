```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# mostramos todos las columnas sin la limitacion de espacio
pd.set_option('display.max_columns', None)
#añadimos el csv al dataFrame
df_sharks = pd.read_csv("attacks.csv", encoding="ISO-8859-1")
```

```python
# Matplotlib inline para visualizar los gráficos de Matplotlib
%matplotlib inline
%config Inlinebackend.figure_format= 'retina'

# Configuración para setear y que todas las fig de Seaborn salgan por defecto con este tamaño
# Se puede especificar el tamaño de cada figura
sns.set_context("poster")
sns.set(rc={"figure.figsize": (12.,6.)})
sns.set_style("whitegrid")
```

```python
copiadfsharks = df_sharks.copy(deep=True)
```

```python
df_sharks
```

```python
copiadfsharks.drop(["Case Number"], axis=1, inplace=True)
copiadfsharks.drop(["pdf"], axis=1, inplace=True)
copiadfsharks.drop(["href formula"], axis=1, inplace=True)
copiadfsharks.drop(["href"], axis=1, inplace=True)
copiadfsharks.drop(["Case Number.1"], axis=1, inplace=True)
copiadfsharks.drop(["Case Number.2"], axis=1, inplace=True)
copiadfsharks.drop(["original order"], axis=1, inplace=True)
copiadfsharks.drop(["Unnamed: 22"], axis=1, inplace=True)
copiadfsharks.drop(["Unnamed: 23"], axis=1, inplace=True)
```

```python
copiadfsharks.dropna(how="all", inplace=True)
```
