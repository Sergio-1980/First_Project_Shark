```python
def find_char(x):
    if re.match("N",x) != None:
        a = re.match("N",x).group(0)
        return a
    elif re.match("Y",x) != None:
        y = re.match("Y",x).group(0)
        return y
    else:
        return None
```

```python
df_sharks2["Fatal(Y/N)"] = df_sharks2["Fatal(Y/N)"].apply(find_char)
```

```python
def continente (x):
    #print (type(x))
    if x in list_Oceania:
        return "Oceania"
    elif x in list_Asia:
        return "Asia"
    elif x in list_Africa:
        return "Africa"
    elif x in list_Europa:
        return "Europe"
    elif x in list_South_America:
        return "South_America"
    elif x in list_North_America:
        return "North_America"
    else:
        return None
```

```python
df_sharks2["Continent"] = df_sharks2["Country"].apply(continente)
```
